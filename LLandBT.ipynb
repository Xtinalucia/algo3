```python
class Node:
    def __init__(self, dataIn = None, next_node = None): #creating nodes, setting, getting and pointing/will use 
        self.data = dataIn                               #these in the functions
        self.nextNode = next_node

    def get_data(self):
        return self.data

    def set_data(self, data):
        self.data = data

    def get_nextNode(self):
        return self.nextNode

    def set_nextNode(self, nextNode):
        self.nextNode = nextNode

class LinkedList: #creates head
    def __init__(self, head = None):
        self.head = head

    def add_Node(self, data):#if no head then assign it to the node then manage the data. compare it to the nodes 
                            #by checking if its < or > to find a place to set it down and point prv to cur and cur 
                            #to next so you dont drop anyhting or its lost
      
        if self.head == None:
            self.head = Node(data)#set bc no head
      
   
        else:
            curr_Node = self.head

            if data < curr_Node.get_data():
                self.head = Node(data, curr_Node)

            else:
                while data > curr_Node.get_data() and curr_Node.get_nextNode() != None:
                    prev_Node = curr_Node
                    curr_Node = curr_Node.get_nextNode()

                if data < curr_Node.get_data():
                    prev_Node.set_nextNode(Node(data, curr_Node))
             
                elif data > curr_Node.get_data() and curr_Node.get_nextNode() == None:
                    curr_Node.set_nextNode(Node(data))#if data is greater than current node and that is the tail
                                                     #then no 'curr_Node' we make it the new tail point to none
    def search(self, data):#traverse when searching, value is the same as the cur node great its true or keep going
        curr_Node = self.head
        while curr_Node != None:
            if data == curr_Node.get_data():
                return True

            else:
                curr_Node = curr_Node.get_nextNode()

        return False
  
    def delete_Node(self, data):#traversing to locate and delete 
        if self.search(data):
            curr_Node = self.head
            if curr_Node.get_data() == data:#equal
                self.head = curr_Node.get_nextNode()#just point to the next one, head drops

            else:
                while curr_Node.get_data() != data:
                    prev_Node = curr_Node
                    curr_Node = curr_Node.get_nextNode()
              
         
                if curr_Node.get_nextNode() != None:#rearrange pointers to drop from middle
                    prev_Node.set_nextNode(curr_Node.get_nextNode())

         
                elif curr_Node.get_nextNode() == None:#set new tail if tail is to be deleted
                     prev_Node.set_nextNode(None)

        else:
             print('Not found')

    def return_as_lst(self):
        lst = []#printing in a organized way adding to a list with append
        curr_Node = self.head
        while curr_Node != None:
            lst.append(curr_Node.get_data())
            curr_Node = curr_Node.get_nextNode()

        return lst

    def size(self):#Counting the number of integers to check the length
        curr_Node = self.head
        count = 0
        while curr_Node != None:
            count += 1
            curr_Node = curr_Node.get_nextNode()
        return count


test1 = LinkedList()
test1.add_Node(20)
test1.add_Node(15)
test1.add_Node(13)
test1.add_Node(14)
test1.delete_Node(14)
test1.delete_Node(17)
print(test1.return_as_lst())
print(test1.size())



```

```python
Binary Search Trees start with a head node 
and each node to the left of that will be smaller, 
each node to the right of it will be greater. 
The far left node should be the lowest number(if one exists) 
that is available. 
The far right node (if one exists) should be the greatest number
```

```python
class Binarytree:#create a root if theres none, check for tail
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

    def addChild(self, data):
        if data == self.data:
            return

        if data < self.data:#check value against head going left if < right when >
            if self.left != None:
                self.left.addChild(data)
            else:
                self.left = Binarytree(data)
        else:
            if self.right:
                self.right.addChild(data)
            else:
                self.right = Binarytree(data)

    def inorder(self):#add to an empty list keeping sm values to left of root greater to the right, combine all
        element = [ ]
    
        if self.left:
            element += self.left.inorder()
    
            element.append(self.data)

        if self.right:
            element += self.right.inorder()
      
        return element

    def search(self,val):#traverse searching for value if its not the root, move through the list, search for 21=T
        if val == self.data:
            return True
        if val < self.data:
            if self.left:
                return self.left.search(val)
            else:
                return False
        else:
            if self.right:
                return self.right.search(val)
            else:
                return False
def buildtree(element):#call the function to build quicker
    root = Binarytree(element[0])
    for i in range(1,len(element)):
            root.addChild(element[i])
    return root
    

element = [39, 87, 21, 42, 95, 52, 12]
tree = buildtree(element)
tree.addChild(1)
print(tree.inorder())
print(tree.search(21))
```
